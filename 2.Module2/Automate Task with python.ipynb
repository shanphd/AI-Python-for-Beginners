### Data Comparision in python 


```python
shan_age = 36
henry_age = 52
Tom_age = 60
sfk_age = 36
```

```python
# Simple Boolean expression
game_age = shan_age < Tom_age
print(game_age)
```

```python

game_age1 = shan_age >= sfk_age
print(game_age1)
game_age2 = shan_age <= Tom_age
print(game_age2)
game_age3 = shan_age == henry_age
print (game_age3)
```

```python
# And Or Boolean Expression 

bool_exp1 = game_age3 and game_age2
print(bool_exp1)
bool_exp2 = game_age3 or game_age2
print (bool_exp2)
```

```python
DataScientist_Job = [{"Company Name" : "Snowflake","job_description":"Need to be Data Scientist","Salary":4000000},
                     {"Company Name": "Tiger Analytics","job_description":"Need to be snowflake and Data scientist","Salary":3900000},
                     {"Company Name": "DTCC","job_description":"Need to be snowflake and Data scientist","Salary":3800000}]
                      
```

```python
print(DataScientist_Job)
```

```python
shan_perfect_company = []
shan_not_perfect_company = []

for salary in DataScientist_Job:
    if salary["Salary"] >= 3900000:
        job = {"Company" : salary["Company Name"],
               "Description" : salary["job_description"],
               "Salary" : salary["Salary"]}
        print(f"Shan you can Apply")
        print(job)
        shan_perfect_company.append(job)
    elif salary["Salary"] <= 3900000:
        job1 = {"Company" : salary["Company Name"],
                "Description" : salary["job_description"],
              "Salary" : salary["Salary"]}
        print(f"Shan Dont Apply")
        print(job1)
        shan_not_perfect_company.append(job1)
```

```python
print(shan_perfect_company)
```

```python
print(shan_not_perfect_company)
```
